```python
height, width = img_copy.shape[:2]
```

```python
mat= np.array([[(0, height),(100, 110), (200, 110), (width, height)]], dtype=np.int32)   
```

```python
zero= np.zeros_like(img_copy)
```

```python
cv2.fillPoly(zero,mat,255)
```

```python
mask=      cv2.fillPoly(zero,mat,255)
```

```python
cv2.bitwise_and(mask,img_copy)
```

```python
plt.imshow(cv2.bitwise_and(mask,img_copy), cmap='gray')
```
